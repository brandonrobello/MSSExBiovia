```python
testdataset_2 = SDFdataset("./data/parm_at_Frosst/zinc.sdf", "./data/antechamber/atomLabels_gaff2.json")
```

```python
print(len(testdataset_2.molecules))
print(len(testdataset_2.Y_labels))
```

```python
featurizer = AtomFeaturizer_RF()

X_features = pd.concat(
    [featurizer.featurize(mol) for mol in testdataset_2.X_molecules.values()],
    ignore_index=True
)

display(X_features)
```

```python
from atoMLtype.RF.RFmodel import RandomForestModel
from sklearn.metrics import accuracy_score, f1_score
```

```python
RFModel = RandomForestModel()

y_labels = testdataset_2.get_labels()

X_test, y_test_decoded = RFModel.train(X_features, y_labels)

y_pred_decoded = RFModel.predict(X_test)

accuracy = accuracy_score(y_test_decoded, y_pred_decoded)
print(f"Random Forest Accuracy: {accuracy}")
```

```python
X_features_wo_neighborhood = pd.concat(
    [featurizer.featurize(mol, include_neighborhood=False) for mol in testdataset_2.X_molecules.values()],
    ignore_index=True
)

display(X_features_wo_neighborhood)
```

```python
RFModel_wo_neighborhood = RandomForestModel()

X_test_wo_neighborhood, y_test_decoded_wo_neighborhood = RFModel_wo_neighborhood.train(X_features_wo_neighborhood, y_labels)

y_pred_decoded_wo_neighborhood = RFModel_wo_neighborhood.predict(X_test_wo_neighborhood)

accuracy_wo_neighborhood = accuracy_score(y_test_decoded_wo_neighborhood, y_pred_decoded_wo_neighborhood)
print(f"Random Forest Accuracy: {accuracy_wo_neighborhood}")
```
