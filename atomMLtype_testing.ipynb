```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter, defaultdict

def categorize_atom_type(atom_type, categories):
    """Returns the elemental category based on the first letter of the atom type."""
    if not atom_type:
        return "unknown"  # Handle empty values
    return atom_type[0].lower() if atom_type[0].lower() in categories else "other"

def generate_elemental_heatmaps(y_test_decoded, y_pred_decoded, min_threshold=3):
    """
    Generates:
    1. Full heatmap of all true vs. predicted pairs (without values).
    2. Heatmaps for (H, C, N, Other) with values and category-level statistics.
    3. A bar chart showing total true values and total mistyped per category.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.
        min_threshold (int): Minimum mistakes per atom type to be included in heatmaps.

    Returns:
        None (Displays all heatmaps and the summary bar chart).
    """
    categories = {"h", "c", "n"}  # "o" is now part of "other"

    # Step 1: Collect True-Prediction Count
    all_matrix = defaultdict(Counter)
    category_mistypes = defaultdict(int)  # Total mistyped counts per category
    category_totals = defaultdict(int)    # Total occurrences per category

    for true_label, pred_label in zip(y_test_decoded, y_pred_decoded):
        true_category = categorize_atom_type(true_label, categories)
        pred_category = categorize_atom_type(pred_label, categories)
        category_totals[true_category] += 1  # Count total occurrences

        if true_label != pred_label:  # Only store mistakes
            all_matrix[true_label][pred_label] += 1
            category_mistypes[true_category] += 1  # Track total misclassifications

    # Convert to DataFrame
    all_df = pd.DataFrame(all_matrix).fillna(0).astype(int)

    # Ensure all labels are present in both dimensions
    all_atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    all_df = all_df.reindex(index=all_atom_types, columns=all_atom_types, fill_value=0)

    # Step 2: Plot Full Heatmap (Without Values)
    true_count = sum(category_totals.values())
    mistype_count = sum(category_mistypes.values())

    plt.figure(figsize=(10, 8))
    sns.heatmap(
        all_df, cmap="Reds", linewidths=0.5, square=True,
        cbar=True, cbar_kws={"label": "Approximate Mistyped Frequency"}, annot=False
    )
    plt.xlabel("Predicted Atom Type")
    plt.ylabel("True Atom Type")
    plt.title(f"Full Heatmap of Mistyped Atom Types (No Values)\nTotal True: {true_count} | Total Mistyped: {mistype_count}")

    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

    # Step 3: Generate Elemental Breakdown Heatmaps
    def plot_heatmap(df, title, true_count, mistype_count):
        """Helper function to plot a heatmap with values in the center."""
        if df.empty:
            print(f"No significant misclassifications for {title}, skipping heatmap.")
            return

        plt.figure(figsize=(8, 6))
        sns.heatmap(
            df, cmap="Reds", linewidths=0.5, square=True,
            cbar=True, cbar_kws={"label": "Approximate Mistyped Frequency"},
            annot=True, fmt="d"  # Display values in the center of the cell
        )
        plt.xlabel("Predicted Atom Type")
        plt.ylabel("True Atom Type")
        plt.title(f"{title}\nTotal True: {true_count} | Total Mistyped: {mistype_count}")
        plt.xticks(rotation=90)
        plt.yticks(rotation=0)
        plt.show()

    # Iterate over each category (H, C, N, Other) and generate heatmaps
    for category in categories | {"other"}:  # Explicitly include "other"
        category_matrix = defaultdict(Counter)

        # 🔹 **Fix: True label determines the heatmap, but ALL predicted labels are included**
        for true_label, pred_dict in all_matrix.items():
            true_category = categorize_atom_type(true_label, categories)
            if true_category == category:
                category_matrix[true_label] = pred_dict  # Keep all predictions

        incorrect_df = pd.DataFrame(category_matrix).fillna(0).astype(int)

        # **Ensure all true and predicted labels are included**
        incorrect_df = incorrect_df.reindex(index=all_atom_types, columns=all_atom_types, fill_value=0)

        # Filter out low-frequency mistakes
        incorrect_df["Total Mistakes"] = incorrect_df.sum(axis=1)
        incorrect_df = incorrect_df[incorrect_df["Total Mistakes"] >= min_threshold]
        incorrect_df = incorrect_df.drop(columns=["Total Mistakes"], errors="ignore")

        # Ensure only relevant columns remain
        relevant_columns = incorrect_df.columns[incorrect_df.sum(axis=0) > 0]
        incorrect_df = incorrect_df[relevant_columns]

        plot_heatmap(
            incorrect_df,
            f"Heatmap of Mistyped {category.upper()} Elements",
            category_totals[category],
            category_mistypes[category]
        )

    # Step 4: Generate Summary Bar Chart
    summary_df = pd.DataFrame.from_dict(
        {"Total True Values": category_totals, "Total Mistyped": category_mistypes}, orient="index"
    ).T.fillna(0).astype(int)

    summary_df.plot(
        kind="bar", figsize=(12, 6), color=["blue", "red"], edgecolor="black"
    )
    plt.title("Total True Values vs. Mistyped Counts per Elemental Category")
    plt.xlabel("Elemental Category")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.legend(title="Category", labels=["Total True Values", "Total Mistyped"])
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report

def evaluate_model(y_test_decoded, y_pred_decoded):
    """
    Computes and plots accuracy and precision for each atom type using scikit-learn.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.

    Returns:
        None (Displays classification metrics and a bar chart).
    """

    # Step 1: Compute Overall Accuracy
    overall_accuracy = accuracy_score(y_test_decoded, y_pred_decoded)
    print(f"Overall Model Accuracy: {overall_accuracy:.4f}\n")

    # Step 2: Compute Per-Class Precision
    atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    precision_scores = precision_score(y_test_decoded, y_pred_decoded, average=None, labels=atom_types)

    # Step 3: Generate Classification Report
    print("Detailed Classification Report:\n")
    print(classification_report(y_test_decoded, y_pred_decoded, labels=atom_types))

    # Step 4: Create DataFrame for Plotting
    performance_df = pd.DataFrame({"Precision": precision_scores}, index=atom_types)
    
    # Step 5: Plot Accuracy and Precision
    plt.figure(figsize=(12, 6))
    performance_df.plot(kind="bar", figsize=(14, 6), color=["blue"], edgecolor="black")

    plt.title("Atom Typing Model Performance: Precision per Atom Type")
    plt.xlabel("Atom Type")
    plt.ylabel("Precision Score")
    plt.xticks(rotation=90)
    plt.ylim(0, 1)
    plt.legend(title="Metric", labels=["Precision"])
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

def plot_label_distributions(y_test_decoded, y_pred_decoded):
    """
    Plots the distribution of true vs. predicted atom labels.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.

    Returns:
        None (Displays a grouped bar chart).
    """

    # Step 1: Count occurrences of true and predicted labels
    true_counts = Counter(y_test_decoded)
    pred_counts = Counter(y_pred_decoded)

    # Step 2: Create a DataFrame
    atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    count_df = pd.DataFrame({
        "True Labels": [true_counts.get(atom, 0) for atom in atom_types],
        "Predicted Labels": [pred_counts.get(atom, 0) for atom in atom_types]
    }, index=atom_types)

    # Step 3: Plot True vs. Predicted Distributions
    count_df.plot(
        kind="bar", figsize=(14, 6), color=["blue", "red"], edgecolor="black", width=0.8
    )

    plt.title("Distribution of True vs. Predicted Atom Labels")
    plt.xlabel("Atom Type")
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.legend(title="Label Type", labels=["True Labels", "Predicted Labels"])
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    
    plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter, defaultdict

def categorize_atom_type(atom_type, categories):
    """Returns the elemental category based on the first letter of the atom type."""
    if not atom_type:
        return "unknown"  # Handle empty values
    return atom_type[0].lower() if atom_type[0].lower() in categories else "other"

def generate_confusion_matrix_heatmap(y_test_decoded, y_pred_decoded):
    """
    Generates a confusion matrix heatmap with:
    - Diagonal elements showing accuracy percentage.
    - Off-diagonal elements showing misclassification rates.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.

    Returns:
        None (Displays heatmap).
    """
    # Step 1: Compute Confusion Matrix
    all_atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    confusion_matrix = pd.DataFrame(0, index=all_atom_types, columns=all_atom_types, dtype=float)

    for true_label, pred_label in zip(y_test_decoded, y_pred_decoded):
        confusion_matrix.loc[true_label, pred_label] += 1

    # Step 2: Normalize the Matrix
    row_sums = confusion_matrix.sum(axis=1)
    normalized_matrix = confusion_matrix.div(row_sums, axis=0) * 100  # Convert to percentages

    # Step 3: Generate Annotations
    annotations = normalized_matrix.copy().astype(str)
    for i in range(len(normalized_matrix)):
        for j in range(len(normalized_matrix.columns)):
            value = normalized_matrix.iloc[i, j]
            if i == j:  # Diagonal elements (accuracy)
                annotations.iloc[i, j] = f"{value:.1f}"
            elif value > 0:  # Off-diagonal elements (misclassification)
                annotations.iloc[i, j] = f"{value:.1f}"

    # Step 4: Plot Heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        normalized_matrix,  fmt="s", cmap="BuPu",
        linewidths=0.5, square=True, cbar=True, cbar_kws={"label": "Percentage"},
        xticklabels=all_atom_types, yticklabels=all_atom_types
    )

    plt.xlabel("Reference Types")
    plt.ylabel("Predicted Types")
    plt.title("Confusion Matrix of Learned vs Reference Atom Type")
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

# # Example dummy data for testing
# y_test_example = ["c", "c", "c1", "c2", "c3", "c3", "hh", "cc", "cp", "cp", "cp"]
# y_pred_example = ["c", "c", "c1", "c2", "c3", "cc", "cc", "cp", "cp", "cc", "cp"]

# # Generate the confusion matrix heatmap
# generate_confusion_matrix_heatmap(y_test_example, y_pred_example)
```

```python
def generate_category_heatmaps_fixed_v3(y_test_decoded, y_pred_decoded):
    """
    Generates category-based confusion matrix heatmaps for atom types.
    - Ensures all predictions that cross into another category are included.
    - Reference types are on the X-axis.
    - H, C, N categories get separate heatmaps, all others go into an 'Other' category.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.

    Returns:
        None (Displays category-wise heatmaps).
    """
    # Define element categories
    categories = {"h", "c", "n"}  # Example categories
    category_labels = {"h": "Hydrogen", "c": "Carbon", "n": "Nitrogen", "other": "Other"}

    # Compute full confusion matrix
    all_atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    confusion_matrix = pd.DataFrame(0, index=all_atom_types, columns=all_atom_types, dtype=float)

    for true_label, pred_label in zip(y_test_decoded, y_pred_decoded):
        confusion_matrix.loc[true_label, pred_label] += 1

    # Normalize to percentages
    row_sums = confusion_matrix.sum(axis=1)
    normalized_matrix = confusion_matrix.div(row_sums, axis=0) * 100  # Convert to percentages
    normalized_matrix = normalized_matrix.fillna(0)  # Handle div-by-zero

    # Generate category heatmaps
    for category in categories | {"other"}:
        category_rows = [label for label in all_atom_types if categorize_atom_type(label, categories) == category]

        if not category_rows:
            continue  # Skip if no relevant data

        # Extract sub-matrix for the category
        category_matrix = normalized_matrix.loc[category_rows, :]

        # Ensure all reference types (X-axis) remain
        relevant_refs = list(category_matrix.index)
        category_matrix = category_matrix.loc[:, relevant_refs]  # Transpose to put reference types on X-axis

        if category_matrix.empty or category_matrix.sum().sum() == 0:
            continue  # Skip empty heatmaps

        # Generate annotations
        annotations = category_matrix.copy().astype(str)
        for i in range(len(category_matrix)):
            for j in range(len(category_matrix.columns)):
                value = category_matrix.iloc[i, j]
                if i == j:  # Diagonal elements (accuracy)
                    annotations.iloc[i, j] = f"{value:.1f}"
                elif value > 0:  # Off-diagonal elements (misclassification)
                    annotations.iloc[i, j] = f"{value:.1f}"

        # Plot Heatmap
        plt.figure(figsize=(12, 10))
        sns.heatmap(
            category_matrix.T, annot=annotations.T, fmt="s", cmap="BuPu",
            linewidths=0.5, square=True, cbar=True, cbar_kws={"label": "Percentage"},
            xticklabels=category_matrix.index, yticklabels=category_matrix.columns,
            annot_kws={"size":8}
        )

        plt.xlabel("Reference Types")
        plt.ylabel("Predicted Types")
        plt.title(f"Confusion Matrix for {category_labels.get(category, 'Unknown')} Atom Types")
        plt.xticks(rotation=45, ha="right")
        plt.yticks(rotation=0)
        plt.show()

# # Example dummy data for testing
# y_test_example = ["c", "c", "c1", "c2", "c3", "c3", "cc", "cc", "cp", "cp", "cp", "h1", "h1", "h2", "n1", "x1", "x2"]
# y_pred_example = ["c", "c", "c1", "c2", "c3", "cc", "cc", "cp", "cp", "cc", "cp", "h1", "h2", "n1", "n1", "x1", "x3"]

# # Generate category-wise heatmaps with the fix
# generate_category_heatmaps_fixed_v3(y_test_example, y_pred_example)
```

```python
def generate_cross_category_heatmap(y_test_decoded, y_pred_decoded):
    """
    Generates a heatmap specifically for cases where the predicted atom type belongs
    to a different elemental category than the reference atom type.

    Args:
        y_test_decoded (list): List of true atom types.
        y_pred_decoded (list): List of predicted atom types.

    Returns:
        None (Displays the cross-category heatmap).
    """
    # Define element categories
    categories = {"h", "c", "n"}  # Example categories
    category_labels = {"h": "Hydrogen", "c": "Carbon", "n": "Nitrogen", "other": "Other"}

    # Compute full confusion matrix
    all_atom_types = sorted(set(y_test_decoded) | set(y_pred_decoded))
    confusion_matrix = pd.DataFrame(0, index=all_atom_types, columns=all_atom_types, dtype=float)

    for true_label, pred_label in zip(y_test_decoded, y_pred_decoded):
        confusion_matrix.loc[true_label, pred_label] += 1

    # Normalize to percentages
    row_sums = confusion_matrix.sum(axis=1)
    normalized_matrix = confusion_matrix.div(row_sums, axis=0) * 100  # Convert to percentages
    normalized_matrix = normalized_matrix.fillna(0)  # Handle div-by-zero

    # Identify cross-category cases
    cross_category_matrix = pd.DataFrame(0, index=all_atom_types, columns=all_atom_types, dtype=float)

    for true_label in all_atom_types:
        true_category = categorize_atom_type(true_label, categories)
        for pred_label in all_atom_types:
            pred_category = categorize_atom_type(pred_label, categories)
            if true_category != pred_category:  # Only keep cross-category misclassifications
                cross_category_matrix.loc[true_label, pred_label] = normalized_matrix.loc[true_label, pred_label]

    # Filter out rows and columns with only zeros
    cross_category_matrix = cross_category_matrix.loc[(cross_category_matrix.sum(axis=1) > 0), :]
    cross_category_matrix = cross_category_matrix.loc[:, (cross_category_matrix.sum(axis=0) > 0)]

    if cross_category_matrix.empty:
        print("No cross-category misclassifications found.")
        return

    # Generate annotations
    annotations = cross_category_matrix.copy().astype(str)
    for i in range(len(cross_category_matrix)):
        for j in range(len(cross_category_matrix.columns)):
            value = cross_category_matrix.iloc[i, j]
            if value > 0:
                annotations.iloc[i, j] = f"{value:.1f}"

    # Plot Heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        cross_category_matrix.T, annot=annotations.T, fmt="s", cmap="BuPu",
        linewidths=0.5, square=True, cbar=True, cbar_kws={"label": "Percentage"},
        xticklabels=cross_category_matrix.index, yticklabels=cross_category_matrix.columns,
        annot_kws={"size": 8}  # Reduce font size for annotations

    )

    plt.xlabel("Reference Types")
    plt.ylabel("Predicted Types")
    plt.title("Cross-Category Misclassification Heatmap")
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.show()

# Example dummy data for testing
y_test_example = ["c", "c", "c1", "c2", "c3", "c3", "cc", "cc", "cp", "cp", "cp", "h1", "h1", "h2", "n1", "x1", "x2"]
y_pred_example = ["c", "c", "c1", "c2", "c3", "cc", "cc", "cp", "cp", "cc", "h1", "h1", "h2", "n1", "x1", "c1"]  # Includes a cross-category misclassification

# Generate the cross-category heatmap
generate_cross_category_heatmap(y_test_decoded, y_pred_decoded)
```
